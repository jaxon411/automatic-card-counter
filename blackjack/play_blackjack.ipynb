```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import pandas as pd
import shoe
import blackjack
import player
```

```python
'''
Plays blackjack manually

PARAMETERS:
bankroll = float of players current dollar count
betsize = int of the baseline bet a player will make every hand

No returns. This is a self-contained method
that plays blackjack with while loops for user input and dealer decisions.
'''
#Play blackjack manually
blackjack.PlayBlackjack(bankroll=500,betsize=5)
```

```python
'''
blackjack.AutomatedBlackjack

Plays automated version of PlayBlackjack.
Uses dictionaries defined in GetInput to employ basic strategy.

PARAMETERS:
nhands = int of number of hands to play and record
counting = boolean to turn card counting on (True) or off (False)
bankroll = float of players current dollar count
betsize = int of the baseline bet a player will make every hand

RETURNS:
final_bankroll = float of the final bankroll the player has
bank_history = a numpy array of the players bankroll every hand
WL_history = a numpy array record of the win-loss for every hand (1=win, 0=loss)
bj_history = a numpy array record of player blackjacks for every hand (1=blackjack, 0=not blackjack)
'''
```

```python
#Blackjack with Card Coutning off
#AutomatedBlackjack parameters: nhands,shoesize,bankroll,counting=False,even_money=False,betsize=5
final_bankroll,bank_history,WL_history,blackjacks_history = blackjack.AutomatedBlackjack(nhands=2500,shoesize=2,counting=False,even_money=True,bankroll=10_000,betsize=13)
```

```python
#Blackjack with Card Coutning on 
#AutomatedBlackjack parameters: nhands,shoesize,bankroll,counting=False,even_money=False,betsize=5
final_bankroll,bank_history,WL_history,blackjacks_history = blackjack.AutomatedBlackjack(nhands=2500,shoesize=2,counting=True,bankroll=10_000,betsize=13)
```

```python
#graph non card counting
bankavg = np.array([])
WL_mean = np.array([])
plt.figure(figsize = (15,10))
for i in range(200):
    final_bankroll,bank_history,WL_history,bj_history = blackjack.AutomatedBlackjack(nhands=2500,shoesize=2,counting=False,even_money=True,bankroll=10_000,betsize=13)
    plt.plot(range(bank_history.shape[0]),bank_history,color = 'blue', alpha=0.1)
    bankavg = np.append(bankavg,[final_bankroll])
    WL_mean = np.append(WL_mean,[WL_history.mean()])


plt.title('Blackjack without counting cards', pad=25, fontsize=25)
plt.ylim(0,20_000)
plt.ylabel('$ Bankroll',rotation=0,labelpad=55,fontsize=15)
plt.xlabel('Number of Hands',labelpad=20,fontsize=15)
plt.axhline(y=10000,linewidth=4, color='black')
plt.text(500, 3500, f'Average Ending Bankroll:\n${round(bankavg.mean())}', horizontalalignment='center',verticalalignment='center',fontsize=20)
plt.text(500, 6000, 'Starting Bankroll:\n$10,000', horizontalalignment='center',verticalalignment='center',fontsize=20)
plt.text(1750, 5000, f'Win/Loss:\n{round(WL_mean.mean()*100,2)}%', horizontalalignment='center',verticalalignment='center',fontsize=20)
plt.show()
```

```python
#graph card counting
bankavg = np.array([])
WL_mean = np.array([])
plt.figure(figsize = (15,10))
for i in range(200):
    final_bankroll,bank_history,WL_history,bj_history = blackjack.AutomatedBlackjack(nhands=2500,shoesize=2,counting=True,bankroll=10_000,betsize=13)
    plt.plot(range(bank_history.shape[0]),bank_history,color = 'red', alpha=0.1)
    bankavg = np.append(bankavg,[final_bankroll])
    WL_mean = np.append(WL_mean,[WL_history.mean()])


plt.title('Blackjack without counting cards', pad=25, fontsize=25)
plt.ylim(0,20_000)
plt.ylabel('$ Bankroll',rotation=0,labelpad=55,fontsize=15)
plt.xlabel('Number of Hands',labelpad=20,fontsize=15)
plt.axhline(y=10000,linewidth=4, color='black')
plt.text(500, 3500, f'Average Ending Bankroll:\n${round(bankavg.mean())}', horizontalalignment='center',verticalalignment='center',fontsize=20)
plt.text(500, 6000, 'Starting Bankroll:\n$10,000', horizontalalignment='center',verticalalignment='center',fontsize=20)
plt.text(1750, 5000, f'Win/Loss:\n{round(WL_mean.mean()*100,2)}%', horizontalalignment='center',verticalalignment='center',fontsize=20)
plt.show()
```
